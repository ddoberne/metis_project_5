```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import time, os
```

```python
def get_script(wiki_table):
    script = []
    base_address = 'https://fireemblem.fandom.com'
    chromedriver = "/Applications/chromedriver" # initiating selenium
    os.environ["webdriver.chrome.driver"] = chromedriver
    driver = webdriver.Chrome(chromedriver)
    for i in wiki_table.find_all('a'):
        print(i['href'])
        time.sleep(5)
        driver.get(base_address + i['href'])
        soup2 = BeautifulSoup(driver.page_source, 'html.parser')
        ps = soup2.find_all('p')
        lis = soup2.find('div', {'id':'content'}).find_all('li')
        for p in ps:
            if p.text[0] != '(':
                script.append(p.text)
        for l in lis:
            if l.text != "":
                if l.text[0] not in "1234567890":
                    script.append(l.text)
    driver.quit()
    return script
```

```python
chromedriver = "/Applications/chromedriver" # initiating selenium
os.environ["webdriver.chrome.driver"] = chromedriver
driver = webdriver.Chrome(chromedriver)
fe_all = 'https://fireemblem.fandom.com/wiki/List_of_Game_Dialogue_Scripts'
driver.get(fe_all)
time.sleep(5)
soup = BeautifulSoup(driver.page_source, 'html.parser')
driver.quit()
```

```python
table1 = soup.find("table", {'class':'wikitable'})
script = []
for i in table1.find_all('a'):
    print(i['href'])
    time.sleep(3)
    driver.get(base_address + i['href'])
    soup2 = BeautifulSoup(driver.page_source, 'html.parser')
    ps = soup2.find_all('p')
    lis = soup2.find('div', {'id':'content'}).find_all('li')
    for p in ps:
        if p.text[0] != '(':
            script.append(p.text)
    for l in lis:
        if l.text[0] not in "1234567890":
            script.append(l.text)
```

```python
df_fe1 = pd.DataFrame(script)
```
